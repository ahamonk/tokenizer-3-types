```python
# Install the library (run once)
!pip install -q tokenizers

from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

# Create a BPE tokenizer
tokenizer = Tokenizer(BPE())

# Split text on whitespace before learning subwords
tokenizer.pre_tokenizer = Whitespace()

# Trainer with a vocabulary size of 5000
trainer = BpeTrainer(
    vocab_size=5000
)

# Train on your text file
tokenizer.train(["A_Midsummer_Nights_Dream.txt"], trainer)

# Print vocabulary size
print("Vocabulary Size:", tokenizer.get_vocab_size())

# Save the tokenizer
tokenizer.save("bpe_tokenizer.json")

print("BPE tokenizer saved successfully!")
```

```python
from tokenizers import Tokenizer

# Load the saved tokenizer
tokenizer = Tokenizer.from_file("bpe_tokenizer.json")

sentence = "The course of true love never did run smooth."

encoding = tokenizer.encode(sentence)

print("Tokens:")
print(encoding.tokens)

print("\nToken IDs:")
print(encoding.ids)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.trainers import WordPieceTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.decoders import WordPiece as WordPieceDecoder

tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))
tokenizer.pre_tokenizer = Whitespace()

# Configure the decoder to use ##
tokenizer.decoder = WordPieceDecoder(prefix="##")

trainer = WordPieceTrainer(
    vocab_size=5000,
    special_tokens=["[UNK]"],
    continuing_subword_prefix="##"
)

tokenizer.train(["A_Midsummer_Nights_Dream.txt"], trainer)

tokenizer.save("wordpiece_tokenizer.json")
```

```python
from tokenizers import Tokenizer
from tokenizers.models import Unigram
from tokenizers.trainers import UnigramTrainer
from tokenizers.pre_tokenizers import Whitespace

tokenizer = Tokenizer(Unigram())
tokenizer.pre_tokenizer = Whitespace()

trainer = UnigramTrainer(vocab_size=5000)

tokenizer.train(["A_Midsummer_Nights_Dream.txt"], trainer)

print("Vocabulary Size:", tokenizer.get_vocab_size())

tokenizer.save("unigram_tokenizer.json")
```
